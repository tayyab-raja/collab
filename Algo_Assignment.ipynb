```
# Insertion sort 
def insertionSort(arr): 
   
    # Traverse from 1 to len(arr) 
    for i in range(1, len(arr)):
        
        key = arr[i] 
        # Move elements of arr[0..i-1], that are 
        # greater than key, to one position ahead 
        # of their current position 
        j = i-1

        while j >=0 and key < arr[j] : 
                arr[j+1] = arr[j] 
                j -= 1

        arr[j+1] = key 
```

```
# Merge Sort
def merge(arr, l, m, r): 
    n1 = m - l + 1
    n2 = r- m 
  
    # create temp arrays 
    L = [0] * (n1) 
    R = [0] * (n2) 
  
    # Copy data to temp arrays L[] and R[] 
    for i in range(0 , n1): 
        L[i] = arr[l + i] 
  
    for j in range(0 , n2): 
        R[j] = arr[m + 1 + j] 
  
    # Merge the temp arrays back into arr[l..r] 
    i = 0     # Initial index of first subarray 
    j = 0     # Initial index of second subarray 
    k = l     # Initial index of merged subarray 
  
    while i < n1 and j < n2 : 
        if L[i] <= R[j]: 
            arr[k] = L[i] 
            i += 1
        else: 
            arr[k] = R[j] 
            j += 1
        k += 1
  
    # Copy the remaining elements of L[], if there 
    # are any 
    while i < n1: 
        arr[k] = L[i] 
        i += 1
        k += 1
  
    # Copy the remaining elements of R[], if there 
    # are any 
    while j < n2: 
        arr[k] = R[j] 
        j += 1
        k += 1
  
# l is for left index and r is right index of the 
# sub-array of arr to be sorted 
def mergeSort(arr,l,r): 
    if l < r: 
  
        # Same as (l+r)//2, but avoids overflow for 
        # large l and h 
        m = (l+(r-1))//2
  
        # Sort first and second halves 
        mergeSort(arr, l, m) 
        mergeSort(arr, m+1, r) 
        merge(arr, l, m, r) 
```

```
import numpy as np
from datetime import datetime

arraysize = []
insertionsorttimetaken = []
mergesorttimetaken = []

for n in (2**p for p in range(6, 15)):
  numbers = np.random.randint(1,10000,n);
  numbers_2 = numbers.copy();
  print(numbers)
  print(numbers_2)


  #Insertion Sort
  start_time = datetime.now()
  insertionSort(numbers) 
  dt = datetime.now() - start_time
  ms = (dt.days * 24 * 60 * 60 + dt.seconds) * 1000 + dt.microseconds / 1000.0
  print (' Time Taken by Inertion Sort to sort '+str(n)+' size of Array is '+str(ms)+' MilliSeconds')
  arraysize.append(n)
  insertionsorttimetaken.append(ms)

  print(numbers)
  print(numbers_2)


  #Merge Sort
  start_time = datetime.now()
  mergeSort(numbers_2,0,n-1) 
  dt = datetime.now() - start_time
  ms = (dt.days * 24 * 60 * 60 + dt.seconds) * 1000 + dt.microseconds / 1000.0
  print (' Time Taken by Merge Sort to sort '+str(n)+' size of Array is '+str(ms)+' MilliSeconds')
  mergesorttimetaken.append(ms)

  print ("Sorted array is:") 
  print(numbers)
  print(numbers_2)
  
```

```
import matplotlib.pyplot as plt

plt.plot(arraysize,insertionsorttimetaken)
plt.xlabel('Array Size')
plt.ylabel('Time Taken Milliseconds')
plt.suptitle('Insertion Sort Graph')
plt.show()

print(insertionsorttimetaken)
print(arraysize)

plt.plot(arraysize,mergesorttimetaken)
plt.xlabel('Array Size')
plt.ylabel('Time Taken Milliseconds')
plt.suptitle('Merge Sort Graph')
plt.show()

print(mergesorttimetaken)
print(arraysize)
```
